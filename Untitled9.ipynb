```python
import datetime

# Kelas Susu untuk data produk
class Susu:
    def __init__(self, nama, harga, stok):
        self.nama = nama
        self.harga = harga
        self.stok = stok

# Inisialisasi daftar menu susu
menu_susu = [
    Susu("Susu Nasional", 1000, 8),
    Susu("Susu Nusantara", 2500, 10),
    Susu("Susu Latte", 5000, 11)
]

keranjang = []
penjualan = []

# Fungsi tampilkan menu utama
def tampilkan_menu():
    print("\nmenu susu nusantara")
    for idx, susu in enumerate(menu_susu, 1):
        print(f"{idx}. {susu.nama.lower()} - Rp{susu.harga}(stok: {susu.stok})")
    print("4. Lihat keranjang pesanan")
    print("5. Checkout")
    print("6. Laporan penjualan")
    print("7. Backup data")
    print("8. Keluar")

# Fungsi tambah ke keranjang
def tambah_ke_keranjang(nomor):
    try:
        susu = menu_susu[nomor - 1]
        if susu.stok > 0:
            keranjang.append(susu)
            susu.stok -= 1
            print(f"âœ… {susu.nama} ditambahin ke keranjang.")
        else:
            print("âŒ Maaf kak, stoknya habis.")
    except IndexError:
        print("âš ï¸ Nomor menu tidak valid.")

# Fungsi tampilkan isi keranjang
def lihat_keranjang():
    if not keranjang:
        print("ğŸ›’ Keranjang masih kosong kak.")
        return
    print("\nğŸ§¾ Isi Keranjang:")
    total = 0
    for i, item in enumerate(keranjang, 1):
        print(f"{i}. {item.nama} - Rp{item.harga}")
        total += item.harga
    print(f"ğŸ’³ Total bayar: Rp{total}")

# Fungsi checkout
def checkout():
    if not keranjang:
        print("ğŸ“­ Belum ada pesanan kak.")
        return
    lihat_keranjang()
    konfirmasi = input("Konfirmasi checkout? (y/n): ").lower()
    if konfirmasi == 'y':
        penjualan.extend(keranjang.copy())
        keranjang.clear()
        print("ğŸ‰ Terima kasih kak! Pesanan kamu sedang diproses.\n")
    else:
        print("ğŸ›‘ Checkout dibatalin.")

# Fungsi laporan penjualan
def laporan_penjualan():
    if not penjualan:
        print("ğŸ“Š Belum ada penjualan kak.")
        return
    print("\nğŸ“ˆ Laporan Penjualan:")
    total = 0
    for i, item in enumerate(penjualan, 1):
        print(f"{i}. {item.nama} - Rp{item.harga}")
        total += item.harga
    print(f"ğŸ’° Total pendapatan: Rp{total}")

# Fungsi backup data ke file
def backup_data():
    waktu = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
    nama_file = f"backup_penjualan_{waktu}.txt"
    with open(nama_file, "w") as file:
        total = 0
        file.write("Laporan Penjualan:\n")
        for item in penjualan:
            file.write(f"{item.nama} - Rp{item.harga}\n")
            total += item.harga
        file.write(f"Total: Rp{total}\n")
    print(f"ğŸ’¾ Data penjualan dibackup ke: {nama_file}")

# Fungsi utama menjalankan aplikasi
def main():
    while True:
        tampilkan_menu()
        pilihan = input("Pilih nomer menu: ")

        if pilihan == '1':
            tambah_ke_keranjang(1)
        elif pilihan == '2':
            tambah_ke_keranjang(2)
        elif pilihan == '3':
            tambah_ke_keranjang(3)
        elif pilihan == '4':
            lihat_keranjang()
        elif pilihan == '5':
            checkout()
        elif pilihan == '6':
            laporan_penjualan()
        elif pilihan == '7':
            backup_data()
        elif pilihan == '8':
            print("makasih banyak sudah mampir kak! Sampai jumpa ğŸ¤—")
            break
        else:
            print("âš ï¸ Pilihan gak valid kak. Coba lagi.")

if __name__ == "__main__":
    main()
```
